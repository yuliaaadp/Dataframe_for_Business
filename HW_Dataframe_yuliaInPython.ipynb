#Read file from Excel file as a main dataset

```python
#read file from excel file
#skiprows untuk menghilangkan baris yang tidak seharusnya terbaca
df = pd.read_excel('/content/drive/MyDrive/DIGITAL_SKOLA/Homework - DataFrame.xlsx', sheet_name = 'Paid-Transaction', skiprows=1)

#menghilangkan kolom 'Unnamed' yang seharusnya tidak terbaca
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df
```

1. In Paid-Transaction sheet, Insert column after Last Name, name the new column as Full Name. Fill in Full Name as i.e. Ayuni . B

```python
#asssign value untuk full-name
#menggabungkan First Name dan Last Name (do slice untuk mengambil inisial berupa huruf pertama)
full_name = df['First Name'].str.title() + " . " + df['Last Name'].str.title().str.slice(0,1)
```

```python
#add new column "Full Name" di index posisi 4 ke tabel df dengan value dari variabel full_name
df.insert(4, 'Full Name', full_name)
df
```

2. In Paid-Transaction sheet, insert two columns next to Paid Date. Name the first one as Paid Date 2 (i.e. 11-Dec-17) and the second as Paid Day (i.e. Mon, Fri).

```python
#import library untuk memudahkan pengolahan data berupa waktu
from datetime import datetime
```

```python
#convert datatype dari kolom Paid Date ke datatype datetime format yyyy-mm-dd
df['Paid Date'] = pd.to_datetime(df['Paid Date'].astype(str), format='%Y-%m-%d')
```

```python
#assign value untuk paid_date
#convert datetime yyyymmdd ke dd-b-yy
paid_date = df['Paid Date'].dt.strftime('%d-%b-%y').str.title()
paid_date.tail(2)
```

```python
#assign value untuk paid_day
#mengambil data berupa hari (3 karakter) dari Paid Date
paid_day = df['Paid Date'].dt.strftime('%a').str.title()
paid_day.tail(2)
```

```python
#add new column "Paid Date 2" di posisi indeks 1 dengan variabel paid_date by insert column
df.insert(1, 'Paid Date 2', paid_date)

#add new column "Paid Day" di posisi indeks 2 dengan variabel paid_day by insert column
df.insert(2, 'Paid Day', paid_day)

df.tail(2)
```

3. In Paid-Transaction sheet, Insert column after Seller Discount, name the new column as GMV. Fill in GMV based on the following Business Rules:

  GMV is Gross Merchandise Value

  GMV contains of Transaction amount substracts Seller Discount but includes Delivery Fee

```python
#assign value untuk gmv berupa formula untuk mendapatkan nilai gmv
#gmv = Transaction amount - Seller Discount + Delivery Fee
gmv = df['Transaction Amount'] + df['Delivery Fee'] - df['Seller Discount']
```

```python
#add new column "GMV" di posisi indeks 11 (setelah kolom Seller Discount) dengan variabel gmv by insert column
df.insert(11, 'GMV', gmv)
```

4. In Paid-Transaction sheet, insert column 'Net Revenue'. 

  Net Revenue = Transaction Fee Amount - Sales Discount - Other Discount

```python
#read excel file untuk mendapatkan tabel referensi Fee Rate
#skiprows untuk menghilangkan baris yang tidak seharusnya terbaca
df1 = pd.read_excel('/content/drive/MyDrive/DIGITAL_SKOLA/Homework - DataFrame.xlsx', sheet_name = 'Transaction Fee', skiprows=4)

#menghilangkan kolom 'Unnamed' yang seharusnya tidak terbaca
df1 = df1.loc[:, ~df1.columns.str.contains('^Unnamed')]
df1.tail(3)
```

```python
#menghapus nilai NaN (di case ini adalah baris yang kosong namun masuk ke dataset)
df1.dropna(inplace = True)
```

```python
#convert datatype dari kolom Month
#int ke datetime format b-yyyy
df1['Month'] = df1['Month'].astype('int')
df1['Month'] = pd.to_datetime(df1['Month'], format = '%Y%m').dt.strftime('%b-%Y')

#convert dari datetime format b-yyyy ke datatype period month
df1['Month'] = pd.to_datetime(df1['Month'], errors='coerce').dt.to_period('m')

df1.head(3)
```

```python
#assign value untuk variabel month
#fungsi untuk mengubah datatype kolom Month dari datetime yyyymmdd ke period month
month = pd.to_datetime(df['Paid Date'], errors='coerce').dt.to_period('m')

#add new column ke main dataset df di indeks posisi 3 dengan variabel month
#kolom Month Period ditambahkan agar ada kolom yang menjadi keys dari tabel df ketika merging dengan df1
df.insert(3, 'Month Period', month)

#sebenarnya bisa langsung convert datatype kolom Paid Date di df menjadi datatype period month
#tapi supaya tidak mengubah kolom yang asli, maka ditambahkan kolom baru
```

```python
#merging main dataset df dengan tabel referensi fee rate df1
#keys pada kolom dengan datatype dan konten yg sama
#copy df dengan hasil merge df dan df1
df = pd.merge(df, df1, how='left', left_on=['Month Period', 'Meta Category'],\
                right_on=['Month', 'Meta Category'])
df.tail(2)
```

```python
#add new column dengan membuat relasi / perhitungan matematis dari kolom yang telah ada
#add column Net Revenue
#Net Revenue = Transaction Fee Amount - Sales Discount - Other Discount
#instead, Transaction Fee Amount = Transaction Fee Rate * Transaction Amount
df['Net Revenue'] = (df['Transaction Fee Rate'] * df['Transaction Amount']) - df['Sales Discount'] - df['Other Discount']

#dengan sintaks di atas maka kolom baru akan ditambahkan di urutan akhir
df.tail(3)
```

5. Create one chart: on primary axis, show monthly GMV figures using bar chart and on secondary axis, plot monthly %Net Revenue as line chart.

  %Net Revenue = Net Revenue / GMV

  Did we do well in December 2017?

```python
#import library untuk visualisasi data
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#add column % Net Revenue by formula
df['% Net Revenue'] = round(df['Net Revenue'] / df['GMV'], 3)
#dfee.tail(2)
```

```python
#count monthly GMV (sum of GMV) by groupby
mgmv = df.groupby(['Month']).agg({'GMV':'sum'})
mgmv
```

```python
#count monthly %Net Revenue (mean of %Net Revenue) by groupby
mnetrev = round(df.groupby(['Month']).agg({'% Net Revenue':'mean'}), 3)
mnetrev
```

```python
#merge hasil monthly gmv dan monthly %Net Revenue untuk dijadikan tabel acuan visualisasi
dmon = mgmv.merge(mnetrev, on='Month').reset_index()
dmon
```

```python
#convert kolom Month ke datatype string agar bisa dijadikan value sumbu x di chart
dmon['Month'] = pd.DataFrame(dmon['Month'].astype(str))
```

```python
#create a chart

#define subplots
fig,ax = plt.subplots()

#add first line to plot
ax.bar(dmon['Month'], dmon['GMV'], color='red', width=0.5, label='GMV')

#add x-axis label
ax.set_xlabel('Month', fontsize=14)

#add y-axis label
ax.set_ylabel('GMV x10^8', color='red', fontsize=16)

#define second y-axis that shares x-axis with current plot
ax2 = ax.twinx()

#add second line to plot
ax2.plot(dmon['Month'], dmon['% Net Revenue'], color='purple', linewidth=3, marker='o', label='% Net Revenue')

#add second y-axis label
ax2.set_ylabel('% Net Revenue', color='purple', fontsize=16)

fig.suptitle('Monthly Progress', fontsize=20)

#show the legend
ax.legend()
ax2.legend()

#show chart
plt.show()
```

Did we do well in December 2017?

Untuk progres di bulan Desember tahun 2017, dilihat dari total penjualan kotor (GMV) terjadi penurunan dari bulan sebelumnya, walaupun tidak signifikan. Sedangkan dari pengamatan % Net Revenue di bulan Desember tersebut terjadi penurunan signifikan bahkan hingga titik minus. Dikarenakan % Net Revenue dalam kalkulasinya merupakan pendapatan bersih, yaitu pendapatan kotor dikurangi dengan biaya-biaya seperti biaya pengiriman, diskon, biaya produksi dan sebagainya, sehingga dari perhitungan ini yang lebih akurat untuk digunakan sebagai parameter perkembangan. Dari hal tersebut dapat disimpulkan bahwa performa penjualan di bulan Desember kurang sehat karena mungkin saja terlalu banyak diskon akhir tahun yang diberikan pada penjualan di bulan ini.

8. Show the minimum, maximum and average number of days taken from payment to confirmation by Meta Category for Orders Paid in November 2017

```python
#read file to get dataset about Confirmed Transaction
#skiprows untuk menghilangkan baris yang tidak seharusnya terbaca
df3 = pd.read_excel('/content/drive/MyDrive/DIGITAL_SKOLA/Homework - DataFrame.xlsx', sheet_name = 'Confirmed-Transaction', skiprows=4)

#menghilangkan kolom 'Unnamed' yang seharusnya tidak terbaca
df3 = df3.loc[:, ~df3.columns.str.contains('^Unnamed')]
df3.tail(2)
```

```python
#merge main dataset df dengan tabel Confirmed-Transaction on Order Number
#untuk mendapatkan informasi order mana saja yang telah berstatus confirmed
#dconf = pd.merge(df3, df, how='left', on=['Order Number'])
dconf = pd.merge(df3, df, how='inner', on=['Order Number'])
dconf.tail(2)
```

```python
#insert new column yang berisi hasil perhitungan interval
#kolom baru Confirmation Period di insert ke urutan indeks 3 dengan value dari variabel conf
dconf.insert(3, 'Confirmation Period', conf)
dconf.tail(2)
```

```python
#filter data agar yang ditampilkan hanya data confirmed transaction on November 2017
dnov = dconf[dconf['Month'] == '2017-11']
dnov.tail(2)
```

```python
#gunakan fungsi pivot untuk melakukan agregasi beruapa rata-rata, minimum dam maximum pada data Confirmation Period
#tampilkan berdasarkan Meta Category
pd.pivot_table(dnov,index = 'Meta Category',values = ['Confirmation Period'] ,aggfunc = ['mean', 'min', 'max'])
```

```python
#atau dengan konsep yang sama, gunakan groupby kemudian juga aggregasi
dnov.groupby('Meta Category').agg({'Confirmation Period': ['mean', 'min', 'max']}).reset_index()
```

9. Calculate %Confirmed Trx and % Cancelled Trx on monthly basis

```python
#read files to get info about Cancelled Transaction
#skiprows untuk menghilangkan baris yang tidak seharusnya terbaca
df4 = pd.read_excel('/content/drive/MyDrive/DIGITAL_SKOLA/Homework - DataFrame.xlsx', sheet_name = 'Cancelled-Transaction', skiprows=1)

#menghilangkan kolom 'Unnamed' yang seharusnya tidak terbaca
df4 = df4.loc[:, ~df4.columns.str.contains('^Unnamed')]
df4.tail(2)
```

```python
#merge main dataset df dengan tabel cancelled transaction on order number
#untuk mendapatkan full data untuk transaksi yang berstatus cancelled
#dcancel = pd.merge(df4, df, how='left', on=['Order Number'])
dcancel = pd.merge(df4, df, how='inner', on=['Order Number'])
dcancel.tail()
```

```python
#buat variabel untuk menghitung jumlah transaksi per bulan
paid = df.groupby('Month')['Order Number'].count()

#buat variabel untuk menghitung monthly % Confirmed Transaction based on monthly Paid Transaction
conf = round((dconf.groupby('Month')['Order Number'].count() / paid * 100), 2)

#buat variabel untuk menghitung monthly % Cancelled Transaction based on monthly Paid Transaction
canc = round((dcancel.groupby('Month')['Order Number'].count() / paid * 100), 2)
```

```python
#create new table that contains the result with dictionary method
transaction = {'#Paid Trx': paid, '% Confirmed Trx': conf, '% Cancelled Trx': canc}
trx = pd.DataFrame(transaction)
trx
```

10. From October to December 2017, which stakeholder had highest cancelled GMV? And what was the main reason?

```python
#filter tabel Cancelled Traansaction, only show Transaction from October to December 2017
dod = dcancel[(dcancel['Month'] > '2017-09')]
dod.tail(3)
```
